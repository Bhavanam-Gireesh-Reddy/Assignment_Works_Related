```python
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Load image from URL
image_url = "https://media.istockphoto.com/id/492583108/photo/cheerful-man-holding-beer-bottle.jpg?s=612x612&w=0&k=20&c=h5CWgeZfw0XBKvF_Wj9KgNUjh15unIn5MHZmhSTqYaQ="  # Replace with your image URL
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
image = np.array(image)

# Display original image
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
from ultralytics import YOLO

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Run detection
results = model(image)

# Extract bottle bounding box
bottle_box = None
for obj in results[0].boxes.xyxy:
    x1, y1, x2, y2 = map(int, obj)
    class_id = int(results[0].boxes.cls[results[0].boxes.xyxy.tolist().index(obj.tolist())])
    if class_id == 39:  # COCO class ID for bottle
        bottle_box = (x1, y1, x2, y2)
        break

if bottle_box:
    print("Bottle detected at:", bottle_box)
else:
    print("No bottle detected.")
```

```python
from segment_anything import SamPredictor, sam_model_registry
import torch

# Load SAM model
sam_checkpoint = "sam_vit_h.pth"
sam = sam_model_registry["vit_h"](checkpoint=sam_checkpoint)
sam.to("cuda" if torch.cuda.is_available() else "cpu")

# Prepare predictor
predictor = SamPredictor(sam)
predictor.set_image(image)

# Get mask using YOLO bounding box
input_box = np.array(bottle_box)
masks, scores, logits = predictor.predict(box=input_box[None, :], multimask_output=True)

# Show mask
plt.imshow(image)
for mask in masks:
    plt.imshow(mask, alpha=0.5)  # Overlay mask
plt.axis("off")
plt.show()
```

```python
# Create mask for inpainting
mask = np.zeros(image.shape[:2], dtype=np.uint8)
cv2.rectangle(mask, (bottle_box[0], bottle_box[1]), (bottle_box[2], bottle_box[3]), 255, -1)

# Remove bottle using inpainting
image_no_bottle = cv2.inpaint(image, mask, 3, cv2.INPAINT_TELEA)

plt.imshow(image_no_bottle)
plt.axis("off")
plt.show()
```

```python
import requests
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
from io import BytesIO
from segment_anything import sam_model_registry, SamPredictor

# Function to load an image from a URL
def load_image_from_url(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# Load the image
image_url = "https://www.agi-glaspac.com/wp-content/uploads/2019/08/750ml-Flint-Bagpiper-Bottle-scaled.jpg"  # Replace with your image URL
image = np.array(load_image_from_url(image_url))

# Load SAM model
sam_checkpoint = "sam_vit_h.pth"
sam = sam_model_registry["vit_h"](checkpoint=sam_checkpoint)
predictor = SamPredictor(sam)

# Detect and segment the bottle
predictor.set_image(image)
masks, scores, _ = predictor.predict(box=None, multimask_output=True)

# Select the best mask
bottle_mask = masks[np.argmax(scores)]

# 🔹 Invert the mask (0s become 1s and vice versa)
inverted_mask = 1 - bottle_mask

# Get bounding box around the bottle
y_indices, x_indices = np.where(inverted_mask)  # Use original mask for bounding box
y_min, y_max = y_indices.min(), y_indices.max()
x_min, x_max = x_indices.min(), x_indices.max()

# Crop the image to only the bottle
cropped_bottle = image[y_min:y_max, x_min:x_max]

# Apply inverted mask to keep only the bottle (removes everything else)
cropped_mask = bottle_mask[y_min:y_max, x_min:x_max]

alpha_channel = ((1 - cropped_mask) * 255).astype(np.uint8)

bottle_rgba = np.dstack([cropped_bottle, alpha_channel])  # Stack RGB + Alpha

bottle_only = cropped_bottle.copy()
bottle_only[cropped_mask == 1] = [0, 0, 0]  # Remove background

# bottle_rgba = np.dstack([bottle_only, np.where(cropped_mask == 0, 255, 0)])  # 255 = opaque, 0 = transparent

# Convert to PIL Image for display and saving
bottle_only_pil = Image.fromarray(bottle_only)

# Convert NumPy array to PIL Image
bottle_transparent = Image.fromarray(bottle_rgba, "RGBA")

# # Display the extracted bottle
plt.figure(figsize=(6, 8))
plt.imshow(bottle_only_pil)
plt.axis("off")
plt.title("Extracted Bottle (With Inverted Mask, Auto-Cropped)")
plt.show()

# Display the extracted bottle with transparency
plt.figure(figsize=(6, 8))
plt.imshow(bottle_transparent)
plt.axis("off")
plt.title("Extracted Bottle (Transparent Background)")
plt.show()

# Save the image (optional)
# bottle_only_pil.save("bottle_only.png")
```

```python
new_bottle = bottle_transparent.resize((bottle_box[2] - bottle_box[0], bottle_box[3] - bottle_box[1]))

# Convert new bottle to NumPy
new_bottle = np.array(new_bottle)

# Extract the region where the bottle should be placed
overlay_region = image_no_bottle[bottle_box[1]:bottle_box[3], bottle_box[0]:bottle_box[2]]

# Ensure new_bottle and overlay_region are the same size
if new_bottle.shape[:2] != overlay_region.shape[:2]:
    new_bottle_pil = Image.fromarray(new_bottle)
    new_bottle_pil = new_bottle_pil.resize((overlay_region.shape[1], overlay_region.shape[0]))
    new_bottle = np.array(new_bottle_pil)

# Extract the alpha channel (transparency) from the new bottle
alpha_mask = new_bottle[:, :, 3] / 255.0  # Normalize to [0, 1]

# Blend the images: original should appear above the new bottle
for c in range(3):  # For each color channel (RGB)
    overlay_region[:, :, c] = (1 - alpha_mask) * overlay_region[:, :, c] + alpha_mask * new_bottle[:, :, c]

# Put the blended region back in the main image
image_no_bottle[bottle_box[1]:bottle_box[3], bottle_box[0]:bottle_box[2]] = overlay_region

# Display the final image
plt.imshow(image_no_bottle)
plt.axis("off")
plt.title("Final Image with Original on Top")
plt.show()
```
